## networkxを使ってグラフの表示

コード確認のため、グラフを表示させる

### ライブラリのインポート

```python
import numpy as np
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
import networkx as nx
import sys
#from scipy.sparse import lil_matrix
#import numpy.linalg as LA
import datetime
#from scipy.sparse import csr_matrix, csc_matrix
#from scipy.sparse.linalg import inv, eigs
import math
from skimage.draw import line
```

### グラフのノード格納

```python
diff = 15

th = 60
    
cnt = 1

now = datetime.datetime.now()
print(now)

print('ノード番号付与中')

#img = cv2.imread('../Input/ORA-663_254-284.png',0)
img = cv2.imread('../Input/test.png',0)

height,width = img.shape

G = nx.Graph()
pos = {}
#edge_weight = []

temp = np.zeros((height,width),dtype = 'i4')

for y in range(height):
    for x in range(width):
        if(img[y,x]>th):
            temp[y,x] = cnt
            
            G.add_node(cnt)
            pos[cnt]=(x,y)
            
            cnt += 1

now = datetime.datetime.now()
print(now)

#print('スパース行列作成')
#A = lil_matrix((cnt+1,cnt+1),dtype='float')
#A = np.zeros((cnt,cnt),dtype='float')

now = datetime.datetime.now()
print(now)

print('隣接行列作成中')

for y in tqdm(range(height)):
    for x in range(width):

        if(img[y,x]>th):
            brig1 = img[y,x]
            node1 = temp[y,x]

            #print(node1)

            if(x!=(width-1)):

                if(img[y,x+1]>th):
                    brig2 = img[y,x+1]
                    diff_n = abs(int(brig1)-int(brig2))
                    node2 = temp[y,x+1]
                    if(diff_n < diff):
                        G.add_edge(node1,node2)
                        #edge_weight.append(diff-diff_n+1)

            if(y!=(height-1)):
                
                if(img[y+1,x]>th):
                    brig2 = img[y+1,x]
                    diff_n = abs(int(brig1)-int(brig2))
                    node2 = temp[y+1,x]
                    if(diff_n < diff):
                        G.add_edge(node1,node2)
                        #edge_weight.append(diff-diff_n+1)

                if(x!=0):
                    if(img[y+1,x-1]>th):
                        brig2 = img[y+1,x-1]
                        diff_n = abs(int(brig1)-int(brig2))
                        node2 = temp[y+1,x-1]
                        if(diff_n < diff):
                            G.add_edge(node1,node2)
                            #edge_weight.append(diff-diff_n+1)

                if(x!=(width-1)):
                    if(img[y+1,x+1]>th):
                        brig2 = img[y+1,x+1]
                        diff_n = abs(int(brig1)-int(brig2))
                        node2 = temp[y+1,x+1]
                        if(diff_n < diff):
                            G.add_edge(node1,node2)
                            #edge_weight.append(diff-diff_n+1)
```

```python
#入力CT画像の番号
img_num = 715

#diff未満の輝度値の差があるノード間を接続

#ノード数カウント用
cnt = 1

#CHECK
check = 0

now = datetime.datetime.now()
print(now)
print('ノード番号付与中')

#入力画像を取得
#img_gray = cv2.imread('../../flower_CT_photo/ORA/[vg-data] ORA/volume_1/ORA-{0:03d}.tif'.format(img_num),0)
#img_gray = cv2.imread('Input/test_g_2.png',0)
#img = cv2.bitwise_not(img_gray)
img_hand = cv2.imread('../Output/Spectral_Clustering_ORA-{0:03d}/ORA-{0:03d}_otsu-hand.png'.format(img_num),0)
img = np.where(img_hand == 255 ,255,0)
#img = np.where(img == 255 ,255,0)

#img_G = cv2.GaussianBlur(img_gray,(5,5),1)
#th, img = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)

#th, img_ = cv2.threshold(img_G, 0, 255, cv2.THRESH_OTSU)

#画像のサイズを取得
height,width = img.shape

G = nx.Graph()
pos = {}
edge_weight = []

#ノード番号を格納する配列を定義
temp = np.zeros((height,width),dtype = 'i4')

#閾値以上のピクセルにノード番号付与
for y in tqdm(range(height)):
    for x in range(width):
        if(img[y,x]>100):
            temp[y,x] = cnt
            
            G.add_node(cnt-1)
            pos[cnt-1]=(x,y)
            
            cnt += 1

now = datetime.datetime.now()
print(now)
print('隣接行列を宣言')

#A = np.zeros((cnt,cnt),dtype='float')

#スパース行列を使うときは以下
#A = lil_matrix((cnt-1,cnt-1),dtype='float')

now = datetime.datetime.now()
print(now)
print('ガボールフィルタ')

#パラメータ設定
k = 30
sigma = 1.1
lam = 10
gam = 0.11

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(0), lam, gam, 0)
img_0 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_0 = img_0.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(15), lam, gam, 0)
img_15 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_15 = img_15.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(30), lam, gam, 0)
img_30 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_30 = img_30.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(45), lam, gam, 0)
img_45 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_45 = img_45.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(60), lam, gam, 0)
img_60 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_60 = img_60.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(75), lam, gam, 0)
img_75 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_75 = img_75.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(90), lam, gam, 0)
img_90 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_90 = img_90.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(105), lam, gam, 0)
img_105 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_105 = img_105.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(120), lam, gam, 0)
img_120 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_120 = img_120.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(135), lam, gam, 0)
img_135 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_135 = img_135.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(150), lam, gam, 0)
img_150 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_150 = img_150.astype('int')

gabor = cv2.getGaborKernel((k,k), sigma, np.radians(165), lam, gam, 0)
img_165 = cv2.filter2D(img, cv2.CV_32S, gabor,cv2.BORDER_CONSTANT)
#img_165 = img_165.astype('int')

#print(img_165.dtype)


now = datetime.datetime.now()
print(now)
print('隣接行列作成中')

r_th = 10
a = 0.001
b = 0.001

for y in tqdm(range(height)):
    for x in range(width):

        if(temp[y,x]>0):
            node1 = temp[y,x]-1
            for y_t in range(y,y+15):
                if(y == y_t):
                    for x_t in range(x,x+15):
                        if((temp[y_t,x_t]>0) and (y != y_t) and (x != x_t)):
                            node2 = temp[y_t,x_t]-1
                            r = math.sqrt((x-x_t)**2 + (y-y_t)**2)
                            if(r < r_th):

                                esc = 0
                                flag = 0
                                i = 0

                                rr,cc = line(y,x,y_t,x_t)

                                while(esc==0):
                                    if(img[rr[i],cc[i]]==0):
                                        flag = 1
                                        esc = 1
                                    elif(i < rr.shape[0]-1):
                                        i += 1
                                    else:
                                        esc = 1

                                if(flag==0):

                                    d = math.sqrt((img_0[y,x]-img_0[y_t,x_t])**2+(img_15[y,x]-img_15[y_t,x_t])**2
                                                  +(img_30[y,x]-img_30[y_t,x_t])**2+(img_45[y,x]-img_45[y_t,x_t])**2
                                                  +(img_60[y,x]-img_60[y_t,x_t])**2+(img_75[y,x]-img_75[y_t,x_t])**2
                                                  +(img_90[y,x]-img_90[y_t,x_t])**2+(img_105[y,x]-img_105[y_t,x_t])**2
                                                  +(img_120[y,x]-img_120[y_t,x_t])**2+(img_135[y,x]-img_135[y_t,x_t])**2
                                                  +(img_150[y,x]-img_150[y_t,x_t])**2+(img_165[y,x]-img_165[y_t,x_t])**2)

                                    W = math.exp(-a*r)+math.exp(-b*d)

                                    #W=math.exp(-a*r)
                                    G.add_edge(node1,node2)
                                    edge_weight.append(W)
                else:
                    for x_t in range(x-15,x+15):
                        if((temp[y_t,x_t]>0) and (y != y_t) and (x != x_t)):
                            node2 = temp[y_t,x_t]-1
                            r = math.sqrt((x-x_t)**2 + (y-y_t)**2)
                            if(r < r_th):

                                esc = 0
                                flag = 0
                                i = 0

                                rr,cc = line(y,x,y_t,x_t)

                                while(esc==0):
                                    if(img[rr[i],cc[i]]==0):
                                        flag = 1
                                        esc = 1
                                    elif(i < rr.shape[0]-1):
                                        i += 1
                                    else:
                                        esc = 1

                                if(flag==0):

                                    d = math.sqrt((img_0[y,x]-img_0[y_t,x_t])**2+(img_15[y,x]-img_15[y_t,x_t])**2
                                                  +(img_30[y,x]-img_30[y_t,x_t])**2+(img_45[y,x]-img_45[y_t,x_t])**2
                                                  +(img_60[y,x]-img_60[y_t,x_t])**2+(img_75[y,x]-img_75[y_t,x_t])**2
                                                  +(img_90[y,x]-img_90[y_t,x_t])**2+(img_105[y,x]-img_105[y_t,x_t])**2
                                                  +(img_120[y,x]-img_120[y_t,x_t])**2+(img_135[y,x]-img_135[y_t,x_t])**2
                                                  +(img_150[y,x]-img_150[y_t,x_t])**2+(img_165[y,x]-img_165[y_t,x_t])**2)

                                    W = math.exp(-a*r)+math.exp(-b*d)

                                    #W=math.exp(-a*r)
                                    G.add_edge(node1,node2)
                                    edge_weight.append(W)
```

### グラフのプロット

```python
#plt.figure(figsize=(20,19),dpi=100)
fig,ax = plt.subplots(figsize=(20,19),dpi=500)
ax.invert_yaxis()

print('G_draw')
nx.draw(G,pos,node_size=0.1,width = edge_weight)
#nx.draw(G,pos,node_size=10)

print('plt_draw')
#plt.show()
#plt.savefig("ORA-test_edge.pdf")
plt.savefig("../Output/ORA-{0:03d}_edge_norm2_gabor.pdf".format(img_num))
```
