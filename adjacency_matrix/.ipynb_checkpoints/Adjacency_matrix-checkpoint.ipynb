```python
%matplotlib inline

import numpy as np
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
import networkx as nx
import sys

#height = 858
#width  = 915

img_num = 479
left = 230
right = 293

height = 59
width  = 63

#depth  = 888
diff = 15

def main():
    
    G = nx.Graph()
    pos = {}
    edge_weight = []
    
    temp = np.zeros((height,width),dtype = 'i4')
    cnt = 0
    
    #img = cv2.imread('../../flower_CT_photo/ORA/[vg-data] ORA/volume_1/ORA-{0:03d}.tif'.format(img_num),0)
    img = cv2.imread('ORA-{0:03d}_{1:d}-{2:d}.png'.format(img_num,left,right),0)
    
    for y in tqdm(range(height)):
        for x in range(width):
            if(img[y,x]>50):
                temp[y,x] = cnt
                G.add_node(cnt)
                pos[cnt]=(x,y)
                cnt += 1
            else:
                img[y,x]=0

    for y in tqdm(range(height)):
        for x in range(width):
            node1 = temp[y,x] 
            if(img[y,x]>0):
                brig1 = img[y,x]
                
                if(x!=(width-1)):
                        
                    if(img[y,x+1]>0):
                        brig2 = img[y,x+1]
                        diff_n = abs(int(brig1)-int(brig2))
                        node2 = temp[y,x+1]
                        if(diff_n < diff):
                            G.add_edge(node1,node2)
                            edge_weight.append(diff-diff_n+1)
                        
                if(y!=(height-1)):
                    
                    if(x!=0):
                        if(img[y+1,x-1]>0):
                            brig2 = img[y+1,x-1]
                            diff_n = abs(int(brig1)-int(brig2))
                            node2 = temp[y+1,x-1]
                            if(diff_n < diff):
                                G.add_edge(node1,node2)
                                edge_weight.append(diff-diff_n+1)
                            
                    if(img[y+1,x]>0):
                        brig2 = img[y+1,x]
                        diff_n = abs(int(brig1)-int(brig2))
                        node2 = temp[y+1,x]
                        if(diff_n < diff):
                            G.add_edge(node1,node2)
                            edge_weight.append(diff-diff_n+1)

                    if(x!=(width-1)):
                        if(img[y+1,x+1]>0):
                            brig2 = img[y+1,x+1]
                            diff_n = abs(int(brig1)-int(brig2))
                            node2 = temp[y+1,x+1]
                            if(diff_n < diff):
                                G.add_edge(node1,node2)
                                edge_weight.append(diff-diff_n+1)
    
    print(nx.info(G))
    
    
    #plt.figure(figsize=(20,19),dpi=100)
    fig,ax = plt.subplots(figsize=(20,19),dpi=100)
    ax.invert_yaxis()
    
    print('G_draw')
    #nx.draw(G,pos,node_size=10,width = edge_weight)
    nx.draw(G,pos,node_size=50)
    
    print('plt_draw')
    #plt.show()
    plt.savefig("ORA-{0:03d}_{1:d}-{2:d}_{3:d}_edge.pdf".format(img_num,left,right,diff))
    #plt.savefig("ORA-{0:03d}_{1:d}.pdf".format(img_num,diff))

if __name__ == "__main__":
    main()
```
